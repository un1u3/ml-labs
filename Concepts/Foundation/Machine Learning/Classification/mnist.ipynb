```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784',as_frame=False)
```

```python
# tells about data 
x = mnist.DESCR
```

```python
X = mnist.data
y = mnist.target
```

```python
y.shape
```

```python
y[0]
```

```python
X_train, X_test, y_train, y_test = X[:6000], X[6000:], y[:6000], y[6000:] 
```

```python
y_test.shape
```

```python
y_train_5 = (y_train == '5')
y_test_5 = (y_test == '5')
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
for i in range(10):
    print(sgd_clf.predict([X[i+12300]]))
```

### Performance Measure 

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
```

```python
cross_val_score(dummy_clf,X_train,y_train_5,cv=3,scoring='accuracy')
```

### Confusion Matrix
The general idea of a confusion matrix is to count the number of times  <br>
instance of class A are classified as Class B, for all A/B pairs <br>
For example,to know the number of times the classifier confused images of 8s with 0s,<br>
you would look at row #8, column #0 of the confusion matrix<br>

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_train_5,y_train_pred)
```
